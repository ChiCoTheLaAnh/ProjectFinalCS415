```python
from pathlib import Path

PROMPT = 'person with red backpack'
CHECKPOINT_DIR = '/content/drive/MyDrive/cv-final-project/checkpoints'
INPUT_IMAGE = '/content/drive/MyDrive/cv-final-project/inputs/example_image.jpg'
INPUT_VIDEO = '/content/drive/MyDrive/cv-final-project/inputs/example_clip.mp4'
OUTPUT_ROOT = '/content/drive/MyDrive/cv-final-project/results/smoke_test'
MAX_FRAMES = 32

Path(OUTPUT_ROOT).mkdir(parents=True, exist_ok=True)
print('Prompt:', PROMPT)
print('Image:', INPUT_IMAGE)
print('Video:', INPUT_VIDEO)
print('Checkpoints:', CHECKPOINT_DIR)
print('Output root:', OUTPUT_ROOT)
```

```python
IMAGE_OUTPUT = f'{OUTPUT_ROOT}/image'

!python3 scripts/run_custom_video.py \
  --config configs/base.yaml \
  --input_video "$INPUT_IMAGE" \
  --prompt "$PROMPT" \
  --output_dir "$IMAGE_OUTPUT" \
  --max_frames 1
```

```python
import json
from IPython.display import Image, display

image_overlay = f'{IMAGE_OUTPUT}/smoke_image_overlay.png'
image_summary = f'{IMAGE_OUTPUT}/run_summary.json'

display(Image(filename=image_overlay))
with open(image_summary, 'r', encoding='utf-8') as handle:
    print(json.dumps(json.load(handle), indent=2))
```

```python
VIDEO_OUTPUT = f'{OUTPUT_ROOT}/video'

!python3 scripts/run_custom_video.py \
  --config configs/base.yaml \
  --input_video "$INPUT_VIDEO" \
  --prompt "$PROMPT" \
  --output_dir "$VIDEO_OUTPUT" \
  --max_frames "$MAX_FRAMES"
```

```python
import json
from IPython.display import Video, display

video_overlay = f'{VIDEO_OUTPUT}/smoke_video_overlay.mp4'
video_summary = f'{VIDEO_OUTPUT}/run_summary.json'

display(Video(video_overlay, embed=True))
with open(video_summary, 'r', encoding='utf-8') as handle:
    print(json.dumps(json.load(handle), indent=2))
```
